```python
#### import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text,d.find_all('em')[-1].text)
```

## 觀察只有取回最近的新聞，在16:12之前的新聞就沒有出現了，使用瀏覽器查看時可以發現資料下拉到頁底時，會再從後端取得較舊的新聞資訊，應該就是AJAX的技術，需要的時候才去後端要資料。
