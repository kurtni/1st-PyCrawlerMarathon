## 作業目標

根據範例：

* 取出今天所有的新聞
* 取出今天下午三點到五點的新聞
* 根據範例，取出三天前下午三點到五點的新聞

```python
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

## 根據範例，取出今天所有的新聞

```python
import time
# 先抓取今天日期
# 每間隔幾秒 自動往下滑
# 確認已經抓取到前一天的新聞，表示完整取得今日新聞
todayTime = time.strftime("%Y/%m/%d", time.gmtime())
print('today=',todayTime)
while True:
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    #browser.execute_script("window.scrollTo(0, 10000000);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    h3Tags= soup.find('div',{'class':'block_content'}).find_all('h3')
    lastTime=h3Tags[-1].span.text[:10]
    #print(h3Tags[-1].span.text)
    if str(todayTime)!=lastTime:
        #print('end',h3Tags[-1].span.text)
        break

for h3Tag in h3Tags:
    if h3Tag.span.text[:10] == todayTime:
        print(h3Tag.span.text,h3Tag.em.text,h3Tag.a.text)
```

## 取出今天下午三點到五點的新聞

```python
for h3Tag in h3Tags:
    if h3Tag.span.text[:10] == todayTime:
        #print(h3Tag.span.text[-5:])
        if h3Tag.span.text[-5:] >= '15:00' and h3Tag.span.text[-5:] <= '17:00' :
            print(h3Tag.span.text,h3Tag.em.text,h3Tag.a.text)
```

## 根據範例，取出三天前下午三點到五點的新聞

```python
# 取得三天前的日期
import datetime
threeDateAgo = datetime.datetime.now() + datetime.timedelta(days=-3)
day = threeDateAgo.strftime("%d")
threeDateAgo = threeDateAgo.strftime("%Y/%m/%d")
print(threeDateAgo,day,threeDateAgo[-2:])
```

```python
# 選取日期後再查詢
from selenium.webdriver.support.ui import Select
selectDay = Select(browser.find_element_by_id("selD"))
selectDay.select_by_value(day)
browser.find_element_by_id('button').click()
```

```python
# 下拉滑鼠，資料出現前一天為止
todayTime = threeDateAgo

while True:
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    #browser.execute_script("window.scrollTo(0, 10000000);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    h3Tags= soup.find('div',{'class':'block_content'}).find_all('h3')
    lastTime=h3Tags[-1].span.text[:10]
    print(h3Tags[-1].span.text)
    if str(todayTime) != lastTime:
        print('end',h3Tags[-1].span.text)
        break
```

```python
# 取出下午三點到五點的新聞
for h3Tag in h3Tags:
    if h3Tag.span.text[:10] == todayTime:
        #print(h3Tag.span.text[-5:])
        if h3Tag.span.text[-5:] >= '15:00' and h3Tag.span.text[-5:] <= '17:00' :
            print(h3Tag.span.text,h3Tag.em.text,h3Tag.a.text)
```
