### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text,d.find_all('em')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
多一個儲存欄位(類別)
'''
emdata=[]
for d in soup.find(class_="part_list_2").find_all('h3'):
    emdata.append(d.find_all('em')[-1].text)
print(emdata)   
```

### ③ 哪一個類別的文章最多

```python
'''
使用list的count函數 做出統計
'''
dict_emdatas = dict()
emsets=set(emdata)
print(emsets)
for emset in emsets:
    dict_emdatas[emset] = emdata.count(emset)
    
print(dict_emdatas)
print(sorted(dict_emdatas.items(), key=lambda d: d[1])[-1])
```
